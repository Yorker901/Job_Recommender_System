```python
#Bar Graph for top 20 Jobs from our dataset
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))                       # set the figure size
plt.title("Top 20 Jobs")                           # giving the title for our graph
t1 = df["Job_Name"].value_counts()                   # get the value counts of the "Job_Name" column
subset = t1[:20]                                  # select the top 20 Job_Name by value count
VR=subset.plot(kind="bar", width=0.8)             # create the bar plot, use barh for horizontal bar chart if required
plt.xticks(fontsize=10)                           # set the x-axis tick labels font size
# add the counts on top of the bars
for i, v in enumerate(subset):
    VR.text(i, v+0.5, str(v), color='red', fontweight='bold', ha='center')
plt.show()
```

```python
#Histogram for Jobs Vs Locations - Top 10
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))  

# select the top 10 Job_Name by count
top_Job_Name = df['Job_Name'].value_counts().head(10).index

# select the top 10 locations by count
top_locations = df['Location'].value_counts().head(10).index

# filter the dataframe to include only the top 10 Job_Name and locations
df_top = df[df['Job_Name'].isin(top_Job_Name) & df['Location'].isin(top_locations)]

# create the histogram
sns.histplot(data=df_top, x='Job_Name', y='Location', color='blue')

# set the title
plt.title('Histogram of Top 10 Jobs vs Top 10 Locations')

# set the axis labels
plt.xlabel('Jobs')
plt.ylabel('Location')

plt.xticks(rotation='vertical', fontsize=10)

# show the plot
plt.show()
```

```python
#Histogram for Job_Name Vs Industry - Top 10
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))  

# select the top 10 Job_Name by count
top_Job_Name = df['Job_Name'].value_counts().head(10).index

# select the top 10  Industry by count
top_Industry = df['Industry'].value_counts().head(10).index

# filter the dataframe to include only the top 10 Job_Name and Industry
df_top = df[df['Job_Name'].isin(top_Job_Name) & df['Industry'].isin(top_Industry)]

# create the histogram
sns.histplot(data=df_top, x='Job_Name', y='Industry', color='blue')

# set the title
plt.title('Histogram of Top 10 Jobs and Top 10 Industry')

# set the axis labels
plt.xlabel('Jobs')
plt.ylabel('Industry')

plt.xticks(rotation='vertical', fontsize=10)

# show the plot
plt.show()
```

```python
#Histogram for Company Vs Locations - Top 10
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))  

# select the top 10 Company by count
top_Company = df['Company'].value_counts().head(10).index

# select the top 10 locations by count
top_locations = df['Location'].value_counts().head(10).index

# filter the dataframe to include only the top 10 Company and locations
df_top = df[df['Company'].isin(top_Company) & df['Location'].isin(top_locations)]

# create the histogram
sns.histplot(data=df_top, x='Company', y='Location', color='blue')

# set the title
plt.title('Histogram of Top 10 Company and Top 10 Locations')

# set the axis labels
plt.xlabel('Companies')
plt.ylabel('Location')

plt.xticks(rotation='vertical', fontsize=10)

# show the plot
plt.show()
```

```python
#Scatterplot for Jobs Vs Location

# select the top 10 job titles by count
top_jobs = df['Job_Name'].value_counts().head(10).index

# create a list of colors for each job title
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan']

# plot each job title and its corresponding locations as a scatter plot
for i, job in enumerate(top_jobs):
    df_job = df[df['Job_Name'] == job]
    plt.scatter(df_job['Location'], df_job['Job_Name'], color=colors[i], label=job)

# set the title and axis labels
plt.title('Scatter Plot of Top 10 Job Titles and Their Locations')
plt.xlabel('Location')
plt.ylabel('Job Title')

# # set the legend
plt.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))

# rotate the x-axis tick labels to vertical
plt.xticks(rotation='vertical')

# show the plot
plt.show()
```
