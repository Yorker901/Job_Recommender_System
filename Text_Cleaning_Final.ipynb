```python
import pandas as pd
df=pd.read_csv("linkedin_data 2023-03-31_final.csv")
```

```python
df.rename(columns={"Loaction":"Location"},inplace=True)
df.rename(columns={"Invovlement":"Type of role"},inplace=True)
df.head(3)
```

```python
df.isnull().sum() #finding if we have any null values
```

```python
df.fillna(0, inplace=True) #filling null values by 0
```

```python
df.isnull().sum() #final replacement we can see there are no null values
```

```python
#Extracting numbers from followers column and adding a new column which is INT type
df['Followers_Numbers'] = df['Followers'].apply(lambda x: int(''.join(filter(str.isdigit, x))))
df.head(2)
```

```python
#Extracting numbers from applicants from Applicant column adding a new column which is INT type
df["New Applicants"] = df["Applicant"].str.extract(r"(\d+)", expand=False)

# Replace NaN values with -1 (or any other default value)
df["New Applicants"].fillna(-1, inplace=True)

# Convert "New Applicants" column to integer data type
df["New Applicants"] = df["New Applicants"].astype(int)

df.head(2)
```

```python
#Extracting numbers from employees from employees column adding a new column which is INT type
import re
df["Employees_Num"] = df["Employees"].apply(lambda x: re.findall(r'\d{1,3}(?:,\d{3})*', x))

# Calculate average for ranges of numbers and round to nearest integer
df["Employees_Avg"] = df["Employees_Num"].apply(lambda x: round(sum(int(num.replace(',', '')) for num in x) / len(x)) if len(x) > 1 else int(x[0].replace(',', '')))

# Convert extracted numbers to integers and take only the first value
df["Employees_Num"] = df["Employees_Num"].apply(lambda x: int(x[0].replace(',', '')))

df.head(2)
```

```python
df=df.drop(columns = ['Date','Employees_Num'])
df.head(2)
```

```python
# Extract city, state, and country using regex patterns
df["City"] = df["Location"].apply(lambda x: re.findall(r'^([^,]+),', x)[0] if re.findall(r'^([^,]+),', x) else None)
df["State"] = df["Location"].apply(lambda x: re.findall(r',\s([^,]+),', x)[0] if re.findall(r',\s([^,]+),', x) else None)
df["Country"] = df["Location"].apply(lambda x: re.findall(r'([^,]+)$', x)[0] if re.findall(r'([^,]+)$', x) else None)
df.head(3)
```

```python
# Clean Type of role column to remove any special characters or values
df["Involvement_1"] = df["Type of role"].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', x))  # Remove special characters
df["Involvement_1"] = df["Involvement_1"].apply(lambda x: x.replace('·', ''))  # Remove dot in between words
# df["Involvement_1"] = df["Type of role"].apply(lambda x: x.replace('-', ' '))  # Replace hyphens with spaces
df.head(4)
```

```python
#Cleaning job type column to remove special characters and adding a new column
df["Job_type_cleaned"] = df["Job_type"].str.replace(r'[^A-Za-z\s]', '').str.replace('-', ' ').str.replace(' ', '')
df.head(1)
```

```python
df.columns
```

```python
df.rename(columns={"Involvement_1":"Type of role cleaned"},inplace=True)
df.head(2)
```

```python
#Lowercasing all the text data

df["Company"] = df["Company"].str.lower()
df["Industry"] = df["Industry"].str.lower()
df["Job_Name"] = df["Job_Name"].str.lower()
df["Job_type_cleaned"] = df["Job_type_cleaned"].str.lower()
df["State"] = df["State"].str.lower()
df["City"] = df["City"].str.lower()
df["Country"] = df["Country"].str.lower()
df["Type of role cleaned"] = df["Type of role cleaned"].str.lower()
df.head(5)
```

```python
#Cleaning job title to remove any special characters
def clean_job_title(title):
    # Remove special characters
    title = ''.join(e for e in title if e.isalnum() or e.isspace())
    # Remove extra spaces
    title = ' '.join(title.split())
    return title

# Apply the clean_job_title function to "Job Title" column
df["Job Title Cleaned"] = df["Job_Name"].apply(clean_job_title)
df.head(2)
```

```python
#Cleaning job title to remove any special characters
def remove_special_characters(text):
    # Use regex to replace any special characters with an empty string
    text = re.sub(r'[^\w\s]', '', text)
    return text

# Convert the Industry column to string before applying the function
df['Industry'] = df['Industry'].astype(str)
# Apply the function to the Industry column
df['Industry'] = df['Industry'].apply(remove_special_characters)
df.head(3)
```

```python
df.to_csv('cleaned_2.csv')
```

```python
#rearranging the columns as per our needs
column_order = ["Job Title Cleaned", "Company", "Industry", "Employees_Avg", "Followers_Numbers",
                "New Applicants", "City", "State", "Country", "Job_type_cleaned", "Type of role cleaned",
                "Job_Name", "Location", "Job_type", "Employees", "Followers", "Applicant","Type of role"]
df = df[column_order]
df.head(5)
```

```python
df=df.drop(columns=["Job_Name", "Location", "Job_type", "Employees", "Followers", "Applicant", "Type of role"])
df.head(10)
```
